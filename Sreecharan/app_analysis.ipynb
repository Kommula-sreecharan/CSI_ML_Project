```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```
url = 'https://raw.githubusercontent.com/SkillsSquad/Machine-Learning/master/DATA/Subcription%20App%20Behaviour.csv'
df = pd.read_csv(url)
df
```

```
df.info()
```

```
df['first_open'] = df['first_open'].apply(lambda x : float(x.split(':')[0]) + float(x.split(':')[1][:2])*60.0 + float(x.split(':')[1][3:])*0)
```

```
df['hour']=pd.factorize(df.hour)[0]
```

```
screen_vocab = set((','.join(df['screen_list'])).split(','))
screen_vocab
```

```
for screen in screen_vocab:
    df[screen] = df.screen_list.str.contains(screen).astype(int)
```

```
df
```

```
df.drop(['user','screen_list','enrolled_date'],axis=1,inplace=True)
```

```
X = df.drop('enrolled', axis = 1)
y = df['enrolled']
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 1)
```

```
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```
from sklearn.metrics import confusion_matrix, accuracy_score
from sklearn.ensemble import RandomForestClassifier
regressor=RandomForestClassifier(n_estimators=175,random_state=1,criterion='gini')
regressor.fit(X_train,y_train)
y_pred_regressor=regressor.predict(X_test)
accuracy_score(y_test, y_pred_regressor)
```

```
from xgboost import XGBClassifier
xgb_model = XGBClassifier()
xgb_model.fit(X_train, y_train)
y_pred_xgb = xgb_model.predict(X_test)
accuracy_score(y_test, y_pred_xgb)
```
